```python
import joblib
import json
import pandas as pd

import nbimporter
# from P1M2_Mujahid import one_hot_encode_and_concat, min_max_scale
```

```python
with open('model.pkl', 'rb') as file_1:
  model = joblib.load(file_1)

with open('scaler.pkl', 'rb') as file_2:
  scaler = joblib.load(file_2)

with open('encoder.pkl', 'rb') as file_3:
  encoder = joblib.load(file_3)

with open('categorical_features.txt', 'r') as file_4:
  categorical_features = json.load(file_4)

with open('selected_cat_features.txt', 'r') as file_5:
  selected_cat_features = json.load(file_5)

with open('non_selected_cat_features.txt', 'r') as file_6:
  non_selected_cat_features = json.load(file_6)

with open('numeric_features.txt', 'r') as file_7:
  numeric_features = json.load(file_7)

with open('selected_num_features.txt', 'r') as file_8:
  selected_num_features = json.load(file_8)

with open('non_selected_num_features.txt', 'r') as file_9:
  non_selected_num_features = json.load(file_9)
```

```python
data = {
    'gender': ['Female', 'Female', 'Male', 'Female', 'Male'],
    'age': [40.0, 59.0, 57.0, 45.0, 50.0],
    'hypertension': [0, 0, 1, 1, 0],
    'heart_disease': [0, 1, 0, 1, 0],
    'ever_married': ['Yes', 'No', 'Yes', 'No', 'Yes'],
    'work_type': ['Private', 'Self-employed', 'Never_worked', 'Govt_job', 'children'],
    'residence_type': ['Urban', 'Urban', 'Urban', 'Rural', 'Urban'],
    'avg_glucose_level': [65.77, 81.64, 217.4, 120.0, 92.5],
    'bmi': [31.2, 32.8, 36.6, 25.0, 28.3],
    'smoking_status': ['never smoked', 'Unknown', 'never smoked', 'formerly smoked', 'smokes']
}

# Membuat DataFrame
df = pd.DataFrame(data)

# Menampilkan DataFrame
df
```

```python
data_enc = df.copy()
data_ohe = encoder.transform(data_enc[selected_cat_features])
feature_names = encoder.get_feature_names_out(selected_cat_features)
data_ohe = pd.DataFrame(data_ohe, columns=feature_names)

new_data = pd.concat([data_enc, data_ohe], axis=1)
new_data.drop(columns=selected_cat_features, inplace=True)
```

```python
new_data[numeric_features] = scaler.transform(new_data[numeric_features])
```

```python
new_data.drop(columns=non_selected_num_features + non_selected_cat_features, inplace=True)
```

```python
df['stroke'] = model.predict(new_data)
```
